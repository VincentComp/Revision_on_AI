```python
import numpy as np
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn import metrics

import pandas as pd

import matplotlib.pyplot as plt

```

```python
from google.colab import drive
drive.mount("/content/drive")
%cd "/content/drive/MyDrive/Colab Notebooks/Self-study/bayes"
%ls  #list directory
```

```python
train_dataset = np.load("train_dataset.npy")
test_dataset = np.load("test_dataset.npy")
train_labels = np.load("train_labels.npy")
test_labels = np.load("test_labels.npy")

print(f'train_dataset : {train_dataset.shape}')
print(f'test_dataset : {test_dataset.shape}')
print(f'train_labels : {train_labels.shape}')
print(f'test_labels : {test_labels.shape}')

pd.DataFrame(train_dataset)
```

```python
#train & fit
bnb = BernoulliNB()
bnb.fit(train_dataset, train_labels)
predictions = bnb.predict(test_dataset); prob = bnb.predict_proba(test_dataset)


#print output
print("Prediction:\n",predictions)

# Print classification report and confusion matrix
print("Classification Report:")
print(classification_report(test_labels, predictions))

print("Confusion Matrix:")
cm = confusion_matrix(test_labels, predictions)
disp = metrics.ConfusionMatrixDisplay(cm)
disp.plot()
plt.show()

```
