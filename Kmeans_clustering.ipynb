```python
def plot_diagram(X, y=None, k=None):
    from itertools import cycle
    cycol = cycle('rgbcmk')                             # Create a color cycle iterable object
    if y is not None:                                   # y is the label of each individual points
        if k is None:                                   # k is number of cluster there is
            k = y.max() + 1                             # The total number of clusters is the max label + 1, assuming
                                                        # the class labels are consecutive and start from 0
        for i, c in zip(range(k), cycol):
            print('label %s: %s' % (str(i), c))
            plt.scatter(*X[y == i].T, c=c)              # Plot points of class with color c
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from random import randint

# Generate 2D data points
X, _ = make_blobs(n_samples=10000, n_features=2,centers=randint(0,10), cluster_std=0.60, random_state=0)

plt.scatter(X[:,0],X[:,1])
plt.show()

```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming that 'X' is your dataset

sse = []
list_k = list(range(1, 10))

for k in list_k:
    km = KMeans(n_clusters=k,random_state=0,n_init = 1, max_iter = 10)
    km.fit(X)
    sse.append(km.inertia_)

# Plot sse against k
plt.figure(figsize=(6, 6))
plt.plot(list_k, sse, '-o')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of squared distance');
```

```python
from sklearn.cluster import KMeans
import numpy as np
kmeans = KMeans(n_clusters=4, random_state=0, n_init = 1,max_iter = 10)
kmeans.fit(X)
Y = kmeans.predict(X)





import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# Assuming 'X' is your 2D dataset and 'y' is the cluster labels

unique_labels = np.unique(Y)
colors = mcolors.ListedColormap(plt.cm.viridis(np.linspace(0, 1, len(unique_labels))))
plt.figure(figsize=(10, 8))

# Scatter the data points
for (label, color) in zip(unique_labels, colors.colors):
    mask = (Y == label)
    plt.scatter(X[mask, 0], X[mask, 1], color=color, label=f'Group {label}', alpha=0.6)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter plot of features colored by cluster label')
plt.legend()
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x')
plt.show()



sse = kmeans.inertia_

print('SSE: ', sse)
```
